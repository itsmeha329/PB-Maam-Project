```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Reading Data.csv')
```

```python


# Choose the column(s) you want to plot
column_to_plot =['SENSOR 1', 'SENSOR 2', 'SENSOR 3', 'SENSOR 4',
                    'SENSOR 5', 'SENSOR 6', 'SENSOR 7', 'SENSOR 8']
# Create the plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed

# Line plot (for numerical data)
plt.plot(df[column_to_plot])
plt.title('Line Plot of ' + column_to_plot)
plt.xlabel('Index')
plt.ylabel(column_to_plot)

# Bar plot (for categorical data)
# plt.bar(df.index, df[column_to_plot])
# plt.title('Bar Plot of ' + column_to_plot)
plt.xlabel('Category')
plt.ylabel(column_to_plot)

# Customize the plot as needed
plt.grid(True)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability

# Show the plot
plt.show()
```

```python


# Select the columns to plot (adjust as needed)
columns_to_plot = ['SENSOR 1', 'SENSOR 2', 'SENSOR 3', 'SENSOR 4',
                    'SENSOR 5', 'SENSOR 6', 'SENSOR 7', 'SENSOR 8']

# Create the figure with subplots
fig, axes = plt.subplots(2, 4, figsize=(15, 8))  # Arrange subplots in a 2x4 grid

# Iterate through columns and create scatter plots
for i, column in enumerate(columns_to_plot):
    ax = axes.flat[i]  # Get the appropriate subplot
    ax.scatter(df.index, df[column], s=5)  # Adjust marker size as needed
    ax.set_title(column)
    ax.set_xlabel('Index')
    ax.set_ylabel(column)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python


# Select the columns to plot (adjust as needed)
columns_to_plot = list(df.columns)  # Plot all columns

# Create the plot
plt.figure(figsize=(12, 6))  # Adjust figure size for better visibility

# Plot each column with a distinct color and label
for i, column in enumerate(columns_to_plot):
    plt.plot(df[column], label=column, alpha=0.7)  # Adjust alpha for transparency

# Customize the plot
plt.title('Plot of All Columns')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
