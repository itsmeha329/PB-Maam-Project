```python
import pandas as pd
from sklearn.cluster import KMeans

# Load your dataset into a DataFrame
data = pd.read_csv("Reading Data.csv")  # Replace with your file path

# Specify the number of clusters (replace with your desired number)
n_clusters = 8

# Create a KMeans model instance
kmeans = KMeans(n_clusters=n_clusters)

# Fit the model to the data
kmeans.fit(data)

# Get the predicted cluster labels for each data point
cluster_labels = kmeans.predict(data)

# Print the cluster labels
print(cluster_labels)

# Access the cluster centers (optional)
cluster_centers = kmeans.cluster_centers_
print(cluster_centers)
```

```python
import matplotlib.pyplot as plt

# Assuming you've already performed clustering and have cluster_labels

# Extract features for plotting
feature1 = data["P"]  # Replace with the name of your first feature
feature2 = data["SENSOR 4"]  # Replace with the name of your second feature

# Create scatter plot with different colors for each cluster
plt.scatter(feature1, feature2, c=cluster_labels)

# Add labels and title
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Cluster Visualization")

# Show the plot
plt.show()
```
