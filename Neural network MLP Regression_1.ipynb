```python
import numpy as np
import pandas as pd

def load_and_preprocess_data():
    """
    Loads and preprocesses sample sensor data and m:p values.

    Returns:
        X_train: Training sensor data.
        y_train: Training m:p values.
        X_test: Test sensor data.
        y_test: Test m:p values.
    """

    # Load data from CSV (replace with your actual data loading)
    data = pd.read_csv('FINALDATA.csv', delimiter=',')
    sensor_data = data.iloc[:, 2:]  # Select all columns from index 2 onwards
    m_p_values = data.iloc[:, 0]  # Select the first column



    # Split data into training and test sets (adjust split ratio as needed)
    split_ratio = 0.8
    num_train_samples = int(split_ratio * len(sensor_data))
    X_train, X_test = sensor_data[:num_train_samples], sensor_data[num_train_samples:]
    y_train, y_test = m_p_values[:num_train_samples], m_p_values[num_train_samples:]

    # Normalize sensor data (replace with your normalization method)
    X_train = (X_train - np.mean(X_train, axis=0)) / np.std(X_train, axis=0)
    X_test = (X_test - np.mean(X_train, axis=0)) / np.std(X_train, axis=0)

    return X_train, y_train, X_test, y_test
```

```python
import tensorflow as tf

# Load and preprocess your data (replace with your actual data loading and preprocessing)
X_train, y_train, X_test, y_test = load_and_preprocess_data()

# Define the MLP model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu'),  # Hidden layer with 64 neurons
    tf.keras.layers.Dense(1)  # Output layer with 1 neuron for the m:p value
])

# Compile the model with appropriate optimizer and loss function
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test))

# Evaluate the model on the test set
test_loss = model.evaluate(X_test, y_test)
print("Test loss:", test_loss)

# Make predictions on new sensor data
new_sensor_data = ...  # Replace with your new sensor data
predicted_m_p_values = model.predict(new_sensor_data)
```
